Exercise 1: Performing Calculations

1. Create two variables, num1 and num2, assigning them integer or float values of your choice.
2. Perform each arithmetic operation (addition, subtraction, multiplication, division) on these variables and print the results.

Exercise 2: Boolean Expressions

1. Assign any two integer values to variables a and b.
2. Write boolean expressions for:
    - a is greater than b
    - a is less than or equal to b
    - a is not equal to b
3. Print the results of these expressions.

Exercise 3: Explore float Precision

1. Assign the value 0.1 + 0.2 to a variable named result.
2. Print result and discuss why it might not be what you expect.
